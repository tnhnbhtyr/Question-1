```python
import pandas as pd
df = pd.read_excel('DailyActivities.xlsx', engine='openpyxl')
df
```

```python
import matplotlib.pyplot as plt

charles = df.plot.pie(y='Charles', figsize=(5,5), autopct='%1.0f%%', labels = df['Area of Interest'])
plt.title('CHARLES')
henry = df.plot.pie(y='Henry', figsize=(5,5), autopct='%1.0f%%', labels = df['Area of Interest'])
plt.title('HENRY')
susan = df.plot.pie(y='Susan', figsize=(5,5), autopct='%1.0f%%', labels = df['Area of Interest'])
plt.title('SUSAN')



plt.show()
```

```python
charles = df.plot.bar(x='Area of Interest', y='Charles', figsize=(5,5))
plt.title('CHARLES')

henry = df.plot.bar(x='Area of Interest', y='Henry', figsize=(5,5))
plt.title('HENRY')
susan = df.plot.bar(x='Area of Interest', y='Susan', figsize=(5,5))
plt.title('SUSAN')


plt.show()
```

```python
compl = df.plot.bar(x='Area of Interest', y=['Charles', 'Henry', 'Susan'], figsize=(15,10))
plt.title('ALL')


plt.show()
```

```python
ax = df.plot(kind='bar', figsize=(15,10))

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height: .0%}', (x + width/2, y + height*1.02), ha='center')
```

```python
ax = df.plot(kind='bar', figsize=(15,10), stacked=True)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height: .0%}', (x + width/2, y + height*1.02), ha='center')
```
